```python
import pandas as pd
```

```python
# Base de datos del IDH 2017 en Perú

idh = pd.read_excel("IDH_PNUD_2017.xlsx")
idh
```

```python
with pd.ExcelFile("./IDH_PNUD_2017.xlsx") as idh:
    df_dep = pd.read_excel(idh, 'Departamental')

df_dep
```

```python
#Seleccionamos las columnas con las variables que usaremos

col_selec = ['Unnamed: 1','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13','Unnamed: 15']

df_col = df_dep[col_selec]

df_col
```

```python
#Le asignamos nombres a las columnas

col_nomb = ['dep','pob','idh','esp_vid','p_sec_com','a_educ','ifam_pc']
df_col.columns = col_nomb

df_col
```

```python
#Eliminamos las filas con valores perdidos

df_sin_nan = df_col.dropna()
df_sin_nan = df_sin_nan.reset_index(drop = True)

df_sin_nan
```

```python
#Eliminamos las primeras dos filas

data_fin = df_sin_nan.drop(0).drop(1)
data_fin = data_fin.reset_index(drop = True)

data_fin
```

```python
import altair as alt
```

## Gráfico 1: IDH por departamento (2017)

```python
alt.Chart(data_fin).mark_point().encode(
    x = alt.X('idh', title = 'Índice de Desarrollo Humano'),
    y = alt.Y('dep', title = 'Departamento')
    ).interactive().properties(
    title={
        "text": ['IDH por departamento (2017)']
    }
)
```

## Gráfico 2: Relación entre secundaria completa e ingreso familiar

```python
alt.Chart(data_fin).mark_point().encode(
    x = alt.X('ifam_pc:Q', title = 'Ingreso familiar per cápita (S/)'),
    y = alt.Y('p_sec_com:Q', title = '% de población con secundaria completa (18 años a más)'),
    color = 'idh',
    tooltip = ['dep','idh']
    ).interactive().properties(
    title={
        "text": ['Relación entre secundaria completa e ingreso familiar'],
        "subtitle": ['Dos de los indicadores que componen el IDH',
                    'Fuente: PNUD (2017)']
    }
)
```
